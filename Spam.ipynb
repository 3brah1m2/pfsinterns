```python
import pandas as pd
spamcsv = pd.read_csv("/content/spam.csv", encoding='latin1')
spamdf = pd.DataFrame(spamcsv)
def spamdetector(df, v1):
  spamlst = []
  for i, v in df[v1].items():
    if v == "spam":
      spamlst.append((i))
  print(spamlst)

spamdetector(spamdf, 'v1')
```
